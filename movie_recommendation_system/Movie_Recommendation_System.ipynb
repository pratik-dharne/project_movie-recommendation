# Data Preprocessing

```python
movies.iloc[0][3]
import ast
```

```python
def convert(obj):
    
    obj=ast.literal_eval(obj)
    
    ll=[i['name'] for i in obj]
    return ll    
        
```

```python
movies['genres']=movies['genres'].apply(convert)
```

```python

movies.head()
```

```python
movies['keywords']=movies['keywords'].apply(convert)
```

```python
movies.head()
```

```python
movies['cast'][0]
```

```python
def convertcast(obj):
    
    obj=ast.literal_eval(obj)
    counter=0
    ll=[]
    for dict in obj:
        if not(counter==3):
            ll.append(dict['name'])
            counter+=1    
        else:
            break
    
    return ll
```

```python
movies['cast']=movies['cast'].apply(convertcast)
```

```python
movies.head()
```

```python
ll=[]
for i in movies['crew'].apply(lambda x:ast.literal_eval(x)):
    for  j in i:
        if j['job']=='Director':
            ll.append(j['name'])
            break
```

```python
def convertcrew(obj):
    
    ll=[]
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            ll.append(i['name'])
            break
            
    return ll
#cannot directly return the list inside the loop because if the obj i.e the crew if not contain the director it will return None
#if return is written inside the if condition as the condition will never be true 
```

```python
movies['crew']=movies['crew'].apply(convertcrew)
```

```python
movies.head()
```

```python
movies['overview']=movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
#removing spaces from selective columns to create tags.
```

```python
movies['genres']=movies['genres'].apply(lambda x : [i.replace(' ','') for i in x])
```

```python
movies['keywords']=movies['keywords'].apply(lambda x:[i.replace(' ','')  for i in x])
```

```python
movies['cast']=movies['cast'].apply(lambda x: [i.replace(' ','')for i in x])
```

```python
movies['crew']=movies['crew'].apply(lambda x: [i.replace(' ','') for i in x])
```

```python
movies.head()
```

```python
movies['tags']=movies['overview']+movies['genres']+movies['keywords']+movies['cast']+movies['crew']
```

```python
movies.head()
```

```python
new_df=movies[['movie_id','title','tags']]
```

```python
new_df
```

```python
#join function convert list element into string 
new_df['tags']=new_df['tags'].apply(lambda x: ' '.join(x))
```

```python
new_df['tags']=new_df['tags'].apply(lambda x:x.lower())
```

```python
new_df.head()
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
    l=[]
    for i in text.split():
        l.append(ps.stem(i))
    return ' '.join(l)
    
    
```

```python
new_df['ntags']=new_df['tags'].apply(lambda x: stem(x)) #applying steaming 
```

```python
new_df.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv= CountVectorizer(max_features=5000,stop_words='english',dtype=np.int32)
```

```python
vectors=cv.fit_transform(new_df['ntags']).toarray()
```

```python
vectors
```

```python
cv.get_feature_names()  #most frequent 5000 words
```

```python
from sklearn.metrics.pairwise import cosine_similarity  #cosine similarity is range from [0 to 1]
```

```python
similarity=cosine_similarity(vectors)     #similarity between movies 
similarity
```

```python
corr=pd.DataFrame(data=cosine_similarity(vectors),columns=new_df['title'].values
,index=new_df['title'].values)
corr      #similarity between movies shown using dataframe movie1 is similar to movie2 0.083462 and so movie1 are 
#similar to all other movies by the data shown in the rows
```

```python
import seaborn as sns
sns.heatmap(corr)
```

```python
#recommend function using array similarity
def recommed(movie):
    
    movie_index = new_df[new_df['title']==movie].index[0]
    distances = similarity[movie_index]
    movies = sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
                                        
    for i in movies:
        print(new_df.iloc[i[0],1])
        
    
```
